````python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
````

````python
df = pd.read_csv('placement.csv')
````

````python
print(df.shape)
df.head()
````

````python
sns.scatterplot(x='cgpa', y='resume_score', hue='placed', style='placed',
                palette='coolwarm', data=df)
plt.title("CGPA vs Resume Score by Placement Status")
plt.show()
````

````python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
````

## 🧩 Explanation of the Code

```python
X = df.iloc[:, 0:2]
y = df.iloc[:, -1]
```

### 📘 Step 1: Selecting Features (`X`)

```python
X = df.iloc[:, 0:2]
```

* `:` → selects **all rows**
* `0:2` → selects **columns with index 0 and 1** (Python excludes the end index `2`)

This means you’re selecting the **first two columns** of the dataset.
If your dataset is about student placements, these columns are likely:

```
cgpa | resume_score
```

✅ Therefore, `X` represents the **feature matrix** — the input data that will be used to predict the outcome.

---

### 📙 Step 2: Selecting Target (`y`)

```python
y = df.iloc[:, -1]
```

* `:` → selects **all rows**
* `-1` → selects the **last column** in the DataFrame

This column is usually the **target variable** — in this case, probably:

```
placed
```

✅ Therefore, `y` represents the **target vector** — the values we want the model to predict.

---

### 🧠 Summary Table

| Variable | Meaning        | Columns Selected                           | Type         |
| -------- | -------------- | ------------------------------------------ | ------------ |
| `X`      | Input features | First two columns (`cgpa`, `resume_score`) | 2D DataFrame |
| `y`      | Target label   | Last column (`placed`)                     | 1D Series    |

---

### 🔍 Checking Shapes

You can confirm the shapes of `X` and `y` using:

```python
print("Shape of X:", X.shape)
print("Shape of y:", y.shape)
```

* `X.shape` → returns `(rows, columns)`
* `y.shape` → returns `(rows,)`

If both have the same number of rows, your data is correctly aligned and ready for model training.


````python
from sklearn.linear_model import Perceptron
p = Perceptron()     
#The Perceptron is one of the earliest and simplest binary classification algorithms in machine learning.
#It’s a type of linear model that tries to find a linear boundary (a straight line or hyperplane) to separate data into two classes.

#🧩 In short: It’s a simple neural network with a single neuron.
````

````python
p.fit(X,y)
````

````python
p.coef_
````

````python
p.intercept_
````

````python
from mlxtend.plotting import plot_decision_regions
````

ℹ️ About mlxtend

mlxtend stands for Machine Learning Extensions.

It provides useful tools that extend scikit-learn — such as:

plot_decision_regions() → visualize decision boundaries of classifiers

SequentialFeatureSelector → feature selection

EnsembleVoteClassifier → combining multiple models

````python
plot_decision_regions(X.values, y.values, clf=p, legend=2)
````

````python
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

# Assuming X, y, and a trained model `p`
plot_decision_regions(X.values, y.values, clf=p, legend=2)
plt.xlabel('CGPA')
plt.ylabel('Resume Score')
plt.title('Decision Boundary for Perceptron')
plt.show()
````
